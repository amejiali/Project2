```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
import numpy as np
from sklearn import preprocessing
import datetime
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
FB_csv = Path("../Resources/FB.csv")
fb_df = pd.read_csv(FB_csv)
fb_df['Date'] = pd.to_datetime(fb_df.Date).dt.date

AAPL_csv = Path("../Resources/AAPL.csv")
appl_df = pd.read_csv(AAPL_csv)
appl_df['Date'] = pd.to_datetime(appl_df.Date).dt.date

AMZN_csv = Path("../Resources/AMZN.csv")
amzn_df = pd.read_csv(AMZN_csv)
amzn_df['Date'] = pd.to_datetime(amzn_df.Date).dt.date

GOOGL_csv = Path("../Resources/GOOGL.csv")
googl_df = pd.read_csv(GOOGL_csv)
googl_df['Date'] = pd.to_datetime(googl_df.Date).dt.date

NFLX_csv = Path("../Resources/NFLX.csv")
nflx_df = pd.read_csv(NFLX_csv)
nflx_df['Date'] = pd.to_datetime(nflx_df.Date).dt.date

predicted_csv = Path("../Resources/PREDICTED_PRICES_JAN_2020.csv")
pp_df = pd.read_csv(predicted_csv)
```

```python
# Set bollinger band window
bollinger_window = 5

# Calculate rolling mean and standard deviation
fb_df['bollinger_mid_band'] = fb_df['Close'].rolling(window=bollinger_window).mean()
fb_df['bollinger_std'] = fb_df['Close'].rolling(window=5).std()

# Calculate upper and lowers bands of bollinger band
fb_df['bollinger_upper_band']  = fb_df['bollinger_mid_band'] + (fb_df['bollinger_std'] * 1)
fb_df['bollinger_lower_band']  = fb_df['bollinger_mid_band'] - (fb_df['bollinger_std'] * 1)

# Calculate bollinger band trading signal
fb_df['bollinger_long'] = np.where(fb_df['Close'] < fb_df['bollinger_lower_band'], 1.0, 0.0)
fb_df['bollinger_short'] = np.where(fb_df['Close'] > fb_df['bollinger_upper_band'], -1.0, 0.0)
fb_df['bollinger_signal'] = fb_df['bollinger_long'] + fb_df['bollinger_short']

fb_df
```
