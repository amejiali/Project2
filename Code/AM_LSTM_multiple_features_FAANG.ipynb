```python
def window_data(df, window, feature_col_number, target_col_number):
    """
    This function accepts the column number for the features (X) and the target (y).
    It chunks the data up with a rolling window of Xt - window to predict Xt.
    It returns two numpy arrays of X and y.
    """
    X = []
    y = []
    for i in range(len(df) - window - 1):
        features = df.iloc[i : (i + window), 1:6].values
        X.append(features)
        target = df.iloc[(i + window), target_col_number]
        y.append(target)
    return np.array(X), np.array(y).reshape(-1, 1)
```

```python
# Creating the features (X) and target (y) data using the window_data() function.
window_size = 5

feature_column = 0
target_column = 0
X_fb, y_fb = window_data(fb_df_scaled, window_size, feature_column, target_column)
#print (f"X_fb sample values:\n{X_fb[:5]} \n")
#print (f"y_fb sample values:\n{y_fb[:5]}")

X_aapl, y_aapl = window_data(aapl_df_scaled, window_size, feature_column, target_column)
#print (f"X_aapl sample values:\n{X_aapl[:5]} \n")
#print (f"y_aapl sample values:\n{y_aapl[:5]}")

X_amzn, y_amzn = window_data(amzn_df_scaled, window_size, feature_column, target_column)
#print (f"X_amzn sample values:\n{X_amzn[:5]} \n")
#print (f"y_amzn sample values:\n{y_amzn[:5]}")

X_nflx, y_nflx = window_data(nflx_df_scaled, window_size, feature_column, target_column)
#print (f"X_nflx sample values:\n{X_nflx[:5]} \n")
#print (f"y_nflx sample values:\n{y_nflx[:5]}")

X_googl, y_googl = window_data(googl_df_scaled, window_size, feature_column, target_column)
#print (f"X_googl sample values:\n{X_googl[:5]} \n")
#print (f"y_googl sample values:\n{y_googl[:5]}")
```

```python
# Use 90% of the data for training and the remainder for testing
split = int(0.90 * len(X_fb))
X_fb_train = X_fb[: split - 1]
X_fb_test = X_fb[split:]
y_fb_train = y_fb[: split - 1]
y_fb_test = y_fb[split:]

X_aapl_train = X_aapl[: split - 1]
X_aapl_test = X_aapl[split:]
y_aapl_train = y_aapl[: split - 1]
y_aapl_test = y_aapl[split:]

X_amzn_train = X_amzn[: split - 1]
X_amzn_test = X_amzn[split:]
y_amzn_train = y_amzn[: split - 1]
y_amzn_test = y_amzn[split:]

X_nflx_train = X_nflx[: split - 1]
X_nflx_test = X_nflx[split:]
y_nflx_train = y_nflx[: split - 1]
y_nflx_test = y_nflx[split:]

X_googl_train = X_googl[: split - 1]
X_googl_test = X_googl[split:]
y_googl_train = y_googl[: split - 1]
y_googl_test = y_googl[split:]
```

```python
# Use the MinMaxScaler to scale data between 0 and 1.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(y_fb)
y_fb_train = scaler.transform(y_fb_train)
y_fb_test = scaler.transform(y_fb_test)

scaler.fit(y_aapl)
y_aapl_train = scaler.transform(y_aapl_train)
y_aapl_test = scaler.transform(y_aapl_test)

scaler.fit(y_amzn)
y_amzn_train = scaler.transform(y_amzn_train)
y_amzn_test = scaler.transform(y_amzn_test)

scaler.fit(y_nflx)
y_nflx_train = scaler.transform(y_nflx_train)
y_nflx_test = scaler.transform(y_nflx_test)

scaler.fit(y_googl)
y_googl_train = scaler.transform(y_googl_train)
y_googl_test = scaler.transform(y_googl_test)
```

```python
X_fb_train = X_fb_train.reshape((X_fb_train.shape[0], X_fb_train.shape[1], 5))
X_fb_test = X_fb_test.reshape((X_fb_test.shape[0], X_fb_test.shape[1], 5))
#print (f"X_train sample values:\n{X_fb_train[:5]} \n")
#print (f"X_test sample values:\n{X_fb_test[:5]}")

X_aapl_train = X_aapl_train.reshape((X_aapl_train.shape[0], X_aapl_train.shape[1], 5))
X_aapl_test = X_aapl_test.reshape((X_aapl_test.shape[0], X_aapl_test.shape[1], 5))
#print (f"X_train sample values:\n{X_aapl_train[:5]} \n")
#print (f"X_test sample values:\n{X_aapl_test[:5]}")

X_amzn_train = X_amzn_train.reshape((X_amzn_train.shape[0], X_amzn_train.shape[1], 5))
X_amzn_test = X_amzn_test.reshape((X_amzn_test.shape[0], X_amzn_test.shape[1], 5))
#print (f"X_train sample values:\n{X_amzn_train[:5]} \n")
#print (f"X_test sample values:\n{X_amzn_test[:5]}")

X_nflx_train = X_nflx_train.reshape((X_nflx_train.shape[0], X_nflx_train.shape[1], 5))
X_nflx_test = X_nflx_test.reshape((X_nflx_test.shape[0], X_nflx_test.shape[1], 5))
#print (f"X_train sample values:\n{X_nflx_train[:5]} \n")
#print (f"X_test sample values:\n{X_nflx_test[:5]}")

X_googl_train = X_googl_train.reshape((X_googl_train.shape[0], X_googl_train.shape[1], 5))
X_googl_test = X_googl_test.reshape((X_googl_test.shape[0], X_googl_test.shape[1], 5))
print (f"X_train sample values:\n{X_googl_train[:5]} \n")
print (f"X_test sample values:\n{X_googl_test[:5]}")
```

```python
# Import required Keras modules
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Define the LSTM RNN model.
model_fb = Sequential()

number_units = 10
dropout_fraction = 0.2

# Layer 1
model_fb.add(LSTM(
    units=number_units,
    return_sequences=True,
    input_shape=(X_fb_train.shape[1], 5))
    )
model_fb.add(Dropout(dropout_fraction))
# Layer 2
model_fb.add(LSTM(units=number_units, return_sequences=True))
model_fb.add(Dropout(dropout_fraction))
# Layer 3
model_fb.add(LSTM(units=number_units))
model_fb.add(Dropout(dropout_fraction))
# Output layer
model_fb.add(Dense(1))

model_aapl = Sequential()

# Layer 1
model_aapl.add(LSTM(
    units=number_units,
    return_sequences=True,
    input_shape=(X_aapl_train.shape[1], 5))
    )
model_aapl.add(Dropout(dropout_fraction))
# Layer 2
model_aapl.add(LSTM(units=number_units, return_sequences=True))
model_aapl.add(Dropout(dropout_fraction))
# Layer 3
model_aapl.add(LSTM(units=number_units))
model_aapl.add(Dropout(dropout_fraction))
# Output layer
model_aapl.add(Dense(1))

model_amzn = Sequential()

# Layer 1
model_amzn.add(LSTM(
    units=number_units,
    return_sequences=True,
    input_shape=(X_amzn_train.shape[1], 5))
    )
model_amzn.add(Dropout(dropout_fraction))
# Layer 2
model_amzn.add(LSTM(units=number_units, return_sequences=True))
model_amzn.add(Dropout(dropout_fraction))
# Layer 3
model_amzn.add(LSTM(units=number_units))
model_amzn.add(Dropout(dropout_fraction))
# Output layer
model_amzn.add(Dense(1))

model_nflx = Sequential()

# Layer 1
model_nflx.add(LSTM(
    units=number_units,
    return_sequences=True,
    input_shape=(X_nflx_train.shape[1], 5))
    )
model_nflx.add(Dropout(dropout_fraction))
# Layer 2
model_nflx.add(LSTM(units=number_units, return_sequences=True))
model_nflx.add(Dropout(dropout_fraction))
# Layer 3
model_nflx.add(LSTM(units=number_units))
model_nflx.add(Dropout(dropout_fraction))
# Output layer
model_nflx.add(Dense(1))

model_googl = Sequential()

# Layer 1
model_googl.add(LSTM(
    units=number_units,
    return_sequences=True,
    input_shape=(X_googl_train.shape[1], 5))
    )
model_googl.add(Dropout(dropout_fraction))
# Layer 2
model_googl.add(LSTM(units=number_units, return_sequences=True))
model_googl.add(Dropout(dropout_fraction))
# Layer 3
model_googl.add(LSTM(units=number_units))
model_googl.add(Dropout(dropout_fraction))
# Output layer
model_googl.add(Dense(1))
```

```python
# Compile the model
model_fb.compile(optimizer="adam", loss="mean_squared_error")
model_aapl.compile(optimizer="adam", loss="mean_squared_error")
model_amzn.compile(optimizer="adam", loss="mean_squared_error")
model_nflx.compile(optimizer="adam", loss="mean_squared_error")
model_googl.compile(optimizer="adam", loss="mean_squared_error")
```
