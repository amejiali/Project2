```python
import alpaca_trade_api as tradeapi
import pandas as pd
import datetime
```

```python
api = tradeapi.REST()
start_date = '2019-01-01T00:00:00-23:59'
end_date = '2019-12-31T00:00:00-23:59'

# Get daily price data for FACEBOOK for 2019.
barset = api.get_barset('FB', 'day', start = start_date, end = end_date)
fb_bars = barset['FB']

# Get daily price data for AMAZON for 2019.
barset = api.get_barset('AMZN', 'day', start = start_date, end = end_date)
amzn_bars = barset['AMZN']

# Get daily price data for APPLE for 2019.
barset = api.get_barset('AAPL', 'day', start = start_date, end = end_date)
aapl_bars = barset['AAPL']

# Get daily price data for NETFLIX for 2019.
barset = api.get_barset('NFLX', 'day', start = start_date, end = end_date)
nflx_bars = barset['NFLX']

# Get daily price data for GOOGLE for 2019.
barset = api.get_barset('GOOGL', 'day', start = start_date, end = end_date)
googl_bars = barset['GOOGL']
```

```python
# Create FACEBOOK's data list from fb_bars 
fb_list = []
for i in range(len(fb_bars)):
    fb_list.append([fb_bars[i].t,
                 fb_bars[i].o,
                 fb_bars[i].h,
                 fb_bars[i].l,
                 fb_bars[i].c,
                 fb_bars[i].v])
    
# Create AMAZON's data list from amzn_bars
amzn_list = []
for i in range(len(amzn_bars)):
    amzn_list.append([amzn_bars[i].t,
                 amzn_bars[i].o,
                 amzn_bars[i].h,
                 amzn_bars[i].l,
                 amzn_bars[i].c,
                 amzn_bars[i].v])

# Create APPLE's data list from aapl_bars
aapl_list = []
for i in range(len(aapl_bars)):
    aapl_list.append([aapl_bars[i].t,
                 aapl_bars[i].o,
                 aapl_bars[i].h,
                 aapl_bars[i].l,
                 aapl_bars[i].c,
                 aapl_bars[i].v])

# Create NETFLIX's data list from nflx_bars
nflx_list = []
for i in range(len(nflx_bars)):
    nflx_list.append([nflx_bars[i].t,
                 nflx_bars[i].o,
                 nflx_bars[i].h,
                 nflx_bars[i].l,
                 nflx_bars[i].c,
                 nflx_bars[i].v])

# Create GOOGLE's data list from googl_bars
googl_list = []
for i in range(len(googl_bars)):
    googl_list.append([googl_bars[i].t,
                 googl_bars[i].o,
                 googl_bars[i].h,
                 googl_bars[i].l,
                 googl_bars[i].c,
                 googl_bars[i].v])
```

```python

#Creating DataFrames from Lists
fb_df = pd.DataFrame(fb_list)
amzn_df = pd.DataFrame(amzn_list)
aapl_df = pd.DataFrame(aapl_list)
nflx_df = pd.DataFrame(nflx_list)
googl_df = pd.DataFrame(googl_list)

cols = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
fb_df.columns=(cols)
amzn_df.columns=(cols)
aapl_df.columns=(cols)
nflx_df.columns=(cols)
googl_df.columns=(cols)

fb_df['Date'] = pd.to_datetime(fb_df.Date).dt.date
amzn_df['Date'] = pd.to_datetime(amzn_df.Date).dt.date
aapl_df['Date'] = pd.to_datetime(aapl_df.Date).dt.date
nflx_df['Date'] = pd.to_datetime(nflx_df.Date).dt.date
googl_df['Date'] = pd.to_datetime(googl_df.Date).dt.date
```
