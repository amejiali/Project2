```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
from sklearn import preprocessing
```

```python
#Reading the .csv files

fb_csv = Path("../Resources/fb_data.csv")
fb_df = pd.read_csv(fb_csv)

aapl_csv = Path("../Resources/aapl_data.csv")
aapl_df = pd.read_csv(aapl_csv)

amzn_csv = Path("../Resources/amzn_data.csv")
amzn_df = pd.read_csv(amzn_csv)

nflx_csv = Path("../Resources/nflx_data.csv")
nflx_df = pd.read_csv(nflx_csv)

googl_csv = Path("../Resources/googl_data.csv")
googl_df = pd.read_csv(googl_csv)
```

```python
#Calculating Volatility = abs(Close - Open) * Volume in a 0-1 Scale

fb_df["Volatility"] = abs(fb_df.Close - fb_df.Open) * fb_df.Volume
x = fb_df.Volatility.values #returns a numpy array
x = x.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled
df = pd.DataFrame(x_scaled)
fb_df["Volatility"] = df

aapl_df["Volatility"] = abs(aapl_df.Close - aapl_df.Open) * aapl_df.Volume
x = aapl_df.Volatility.values #returns a numpy array
x = x.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled
df = pd.DataFrame(x_scaled)
aapl_df["Volatility"] = df

amzn_df["Volatility"] = abs(amzn_df.Close - amzn_df.Open) * amzn_df.Volume
x = amzn_df.Volatility.values #returns a numpy array
x = x.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled
df = pd.DataFrame(x_scaled)
amzn_df["Volatility"] = df

nflx_df["Volatility"] = abs(nflx_df.Close - nflx_df.Open) * nflx_df.Volume
x = nflx_df.Volatility.values #returns a numpy array
x = x.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled
df = pd.DataFrame(x_scaled)
nflx_df["Volatility"] = df

googl_df["Volatility"] = abs(googl_df.Close - googl_df.Open) * googl_df.Volume
x = googl_df.Volatility.values #returns a numpy array
x = x.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled
df = pd.DataFrame(x_scaled)
googl_df["Volatility"] = df
```
